```python
import pandas as pd
```

```python
df = pd.read_csv('salario_prof.csv')
df
```

```python
list(df.columns)
```

```python
#O estado com menor salário
menor_salario = df.loc[df['Salario']==df['Salario'].min()] #trazendo o menor salário
print("O estado com menor salário é: ")
menor_salario
```

```python
#O estado com maior salário
menor_salario = df.loc[df['Salario']==df['Salario'].max()] #trazendo o menor salário
print("O estado com maior salário é: ")
menor_salario
```

```python
#Posição em que o estado de Goiás está
# Criando ranking
df['ranking'] = df['Salario'].rank(ascending=False, method='min')

# Pegando o ranking de Goiás
posicao = df.loc[df['UF'] == 'GO', 'ranking'].values[0]

# Print correto
print(f"O estado de Goiás ocupa a {int(posicao)}º posição no ranking de salários")

# Exibindo os dados de GO
df[df['UF'] == 'GO']
```

```python
#A média salárial entre todos os estados
media = df['Salario'].sum() / df['Salario'].count()
print(f"A média salarial dos professores no Brasil é: {media:.2f}")
```

```python
#Mostrar a média salárial entre todos os estados por região
media_regiao = df.groupby('Regiao')['Salario'].mean()
print("A média salarial por região brasileira:")
print(media_regiao.round(2))
```

```python
#Mostrar o maior salário por região
maior_regiao = df.groupby('Regiao')['Salario'].max()

regiao_maior = maior_regiao.idxmax()
valor = maior_regiao.max()

print(f"A região com maior salário é {regiao_maior}, com R$ {valor:.2f}")
```

```python
#Mostrar o menor salário por região
menor_regiao = df.groupby('Regiao')['Salario'].min()

regiao_menor = menor_regiao.idxmin()
valor = menor_regiao.min()

print(f"A região com maior salário é {regiao_menor}, com R$ {valor:.2f}")
```

```python
#Mostrar a média salarial por região
media_regiao = df.groupby('Regiao')['Salario'].mean()

regiao_media = media_regiao.idxmax()
valor_media = media_regiao.max()

print("Média salarial por região: ")
media_regiao.round(2)
```

```python
#Mostrar a região com maior e menor média salarial
media_salarial_por_regiao = df.groupby('Regiao')['Salario'].mean().reset_index()

#encontrando regiao com a maior media
regiao_maior_media = media_salarial_por_regiao.loc[media_salarial_por_regiao['Salario'].idxmax()]

#encontrando regiao com a maior media
regiao_menor_media = media_salarial_por_regiao.loc[media_salarial_por_regiao['Salario'].idxmin()]

#exibir os resultados
print(f"A região com a maior média é: {regiao_maior_media['Regiao']} com um salário médio de R$ {regiao_maior_media['Salario']:.2f}")
print(f"A região com a menor média é: {regiao_menor_media['Regiao']} com um salário médio de R$ {regiao_menor_media['Salario']:.2f}")
```

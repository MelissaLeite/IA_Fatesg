```python
import pandas as pd 
import numpy as np
```

```python
u = pd.Series([1, 3, 5, 6, 8])
print(u)
```

```python
v = pd.Series()
print(v)
```

```python
data = np.array([1, 3, 5, 7, 9])
m = pd.Series(data)
m
```

```python
dados = np.array(['a', 'b', 'c', 'd', 'e'])
k = pd.Series(dados, index=[100, 101, 102, 103, 104])
k
```

```python
#dicionário em python
num = {'a' : 0., 'b' : 1., 'c' : 2.}
s = pd.Series(num)
s
```

```python
idades = {'Melissa': 28, 'Maria': 30, 'Pedro':25, 'Ana': 18}
i = pd.Series(idades)
i
```

```python
data = {'a': 0., 'b': 1., 'c': 2.}
p = pd.Series(data,index=['b', 'c', 'd,', 'a'])
print(p)
```

```python
#imprimir o mesmo valor em todos os índices
w = pd.Series(5, index=[0, 1, 2, 3])
print(w)
```

```python
#Acessando dados da série por posição
d = pd.Series([1,2,3,4,5], index = ['a', 'b', 'c', 'd', 'e'])
print(d)

#buscando o primeiro elemento
print('O primeiro elemento é o: ', d[2])
```

```python
#buscando os 3 primeiros elementos
print(d[:3])
```

```python
#buscando os 3 últimos elementos
print(d[-3:])
```

```python
#buscando valores entre a lista (sempre busca um a menos (intervalo aberto))
print(d[2:4])
```

```python
#Recuperar dados usando rótulo (Index)
print(d['a'])
```

```python
# Mostra na tela os valores correspondentes aos índices 'a', 'c' e 'd'.
print(d[['a', 'c', 'd']])
```

```python
#Se um rótulo não existir, uma exceção será gerada:
print(d['f'])
```

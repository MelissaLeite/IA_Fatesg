```python
'''
Nível 5: Estrutura Multidimensional (DataFrame). 
O DataFrame é uma estrutura bidimensional (tabular) de tamanho mutável. De acordo com o que foi 
apresentado até a página 25, ele pode ser visualizado como uma tabela SQL ou planilha. Escreva a 
sintaxe do construtor completo do DataFrame, listando seus 5 parâmetros principais 
(data, index, columns, dtype, copy) .
'''
```

```python
import pandas as pd

dados = [['Melissa', 28], ['Vitória', 23], ['Karem', 25]] #lista de lista
df = pd.DataFrame(
    dados, 
    columns=['nome', 'idade'], #nomeando as colunas
    index=['a', 'b', 'c'], #nomeando as linhas
    dtype=None,
    copy=True
)
print(df)
```

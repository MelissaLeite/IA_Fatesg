```python
'''
Nível 4: Uso de Dicionários e Tratamento de Dados Ausentes (NaN). 
Se passarmos um dicionário para o construtor pd.Series() e especificarmos um índice com uma chave que não existe no dicionário original, 
o Pandas preencherá o valor com NaN (Not a Number).
Crie um dicionário: {'jan': 100, 'fev': 200}.
Crie uma Series a partir dele, mas use o índice: ['jan', 'fev', 'mar'].
Imprima o resultado e observe o que acontece com o mês de 'mar'.
'''
```

```python
import pandas as pd

print(type(pd))
print(type(pd.Series))

dic = {'jan': 100, 'fev': 200}
meses = pd.Series(dic, index=['jan', 'fev', 'mar'])

print(meses)
```

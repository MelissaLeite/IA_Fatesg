```python
'''
Nível 3: Manipulação de Índices Personalizados. Ao criar uma Series, podemos definir rótulos personalizados 
para os eixos. Crie uma Series utilizando um array Numpy com os valores [10, 20, 30, 40], mas defina 
explicitamente os índices como ['A', 'B', 'C', 'D']. Imprima a Series resultante.
'''
```

```python
import pandas as pd
numeros = pd.Series ([10, 20, 30, 40], index = ['A', 'B', 'C', 'D'])
print(numeros)
```
